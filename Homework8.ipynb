```python
class Media:
    def play(self):
        return "Media play"

class Song(Media):
    def play(self):
        return "Playing audio.."

class Movie(Media):
    def play(self):
        return "Playing video at 1080p.."

class Podcast(Media):
    def play(self):
        return "Playing podcast..."

def play_all(Media):
    for media in Media:
        print(media.play())
media_list = [Song(), Movie(), Podcast()]
play_all(media_list)
```
