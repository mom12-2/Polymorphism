```python
class Media:
    def play(self):
        return 'Media play'
class Song(Media):
    def play(self):
        return "Playing audio.."
class Movie(Media):
    def play(self):
        return "Playing video at 1080p.."
class Podcast(Media):
    def play(self):
        return "Playing podcast..."
Media_list = [Song(), Movie(), Podcast()]
for Media in Media_list:
    print(Media.play())
    
```
